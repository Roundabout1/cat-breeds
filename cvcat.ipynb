```python
import tensorflow as tf
model = tf.keras.models.load_model('my_model.h5')
```

```python
import json

with open('data/id_to_breed.json', 'r') as f:
    raw_id_to_breed = json.load(f)

id_to_breed = {}
for str_id in raw_id_to_breed:
    id_to_breed[int(str_id)] = raw_id_to_breed[str_id]
print(f"The dictionary is {id_to_breed}")
```

```python
import numpy as np
import cv2

def predict_breed(model, img):
    resized_img = cv2.resize(img, (299, 299))
    resized_img = resized_img[...,::-1]
    resized_img = np.expand_dims(resized_img, axis=0)
    predicted = model.predict(resized_img)
    return id_to_breed[np.argmax(predicted)+1].replace('_', ' ')
```

```python
import cv2

face_cascade = cv2.CascadeClassifier('data/haarcascade_frontalcatface.xml')  

cap = cv2.VideoCapture('data/british.mp4')

skip = True

while cap.isOpened():
    skip = not skip 
    ret, img = cap.read()

    if not ret:
        break
    
    if skip:
        continue
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
  
    faces = face_cascade.detectMultiScale(gray)  
    
    font = cv2.FONT_HERSHEY_SIMPLEX

    font_scale = 1

    color = (255, 0, 0)

    thickness = 2

    for (x,y,w,h) in faces:  
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,255,0),2)  
        roi_gray = gray[y:y+h, x:x+w]  
        roi_color = img[y:y+h, x:x+w] 
        frame = img[y:y+h, x:x+w]
        breed_name = predict_breed(model, frame)
        text_size = cv2.getTextSize(breed_name, font, font_scale, thickness)
        text_x = int((x+w) / 2)
        text_y = int((y+h) / 2)
        cv2.putText(img, breed_name, (text_x, text_y), font, font_scale, color, thickness)
        print(breed_name)
  
  
    cv2.imshow('img',img)  
    
    # при нажатии на esc, видео закрывается 
    k = cv2.waitKey(30) & 0xff
    if k == 27:  
        break

cap.release()
cv2.destroyAllWindows()
```
